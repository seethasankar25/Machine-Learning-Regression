```python
import pandas as pd
```

```python
dataset=pd.read_csv("Salary_Data.csv")#read individual column
```

```python
dataset
```

```python
independent=dataset[["YearsExperience"]] #read individual column and assign it to variable
```

```python
independent
```

```python
dependent=dataset[["Salary"]] #read individual column and assign it to variable
```

```python
#The sklearn has libraries for AI models and we are importing train_test_split model to split the dataset for training and testing
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#The sklearn has libraries to select specific linear regression models for training 
from sklearn.linear_model import LinearRegression
#assigning LinearRegression class to variable regressor
regressor=LinearRegression()
#Feeding the training data to the linear regression model
regressor.fit(X_train,y_train)
```

```python
#finding the weight/gradient/slope of the model
weight=regressor.coef_
weight
```

```python
#finding the bias/intercept/origin for the model
bias=regressor.intercept_
bias
```

```python
#Calling predict function and passing input test dataset, assign this to variable y_pred
y_pred=regressor.predict(X_test)
```

```python
#Finding the R-squared value of our model by comparing the output test dataset and predicted dataset
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
#the R squared score close to one defines a good model
r_score
```

```python
#we using pickle libraries to save the model
import pickle
filename="finalized_model_linear.sav"
```

```python
#write the binaries into the file
pickle.dump(regressor,open(filename,'wb'))
```

```python
#load the model from the file
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
```

```python
#check the result by pssing an input for prediction
#input as 13 years of service provides a predicted output for expected salary
result=loaded_model.predict([[13]])
result
```
